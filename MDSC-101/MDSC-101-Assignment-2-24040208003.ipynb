## b)Read a user defined matrix. Verify the existence of inverse and compute its inverse.

```python
def input_matrix(rows, cols):
    matrix = []
    print(f"Enter the elements of a {rows}x{cols} matrix row by row with spaces:")
    for i in range(rows):
        row = list(map(int, input().split()))
        if len(row) != cols:
            raise ValueError(f"Row {i+1} does not have {cols} elements")
        matrix.append(row)
    return matrix

def determinant(matrix):
    if len(matrix) == 0:
        return 0
    elif len(matrix) == 1:
        return matrix[0][0]
    elif len(matrix) == 2:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
    else:
        det = 0
        for i in range(len(matrix)):
            submatrix = [row[:i] + row[i+1:] for row in matrix[1:]]
            det += (-1) ** i * matrix[0][i] * determinant(submatrix)
        return det

def get_inverse(matrix):
    n = len(matrix)
    det = determinant(matrix)
    if det == 0:
        return None  # Inverse does not exist

    # Calculate the adjugate matrix
    adj_matrix = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            submatrix = [row[:j] + row[j+1:] for row in matrix[:i] + matrix[i+1:]]
            adj_matrix[j][i] = (-1) ** (i + j) * determinant(submatrix)

    # Calculate the inverse matrix
    inverse_matrix = [[element / det for element in row] for row in adj_matrix]
    return inverse_matrix

rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

if rows != cols:
    print("Inverse can only be calculated for square matrices.")
else:
    matrix = input_matrix(rows, cols)
    inverse = get_inverse(matrix)
    if inverse is None:
        print("The inverse of the given matrix does not exist (determinant is 0).")
    else:
        print("The inverse of the given matrix is:")
        for row in inverse:
            print(row)
```
