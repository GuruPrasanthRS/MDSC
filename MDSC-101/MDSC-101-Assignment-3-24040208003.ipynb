Write a code using basic python (without using library functions) to find the rank of a user
defined matrix.

```python
def matrix_rank(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    rank = min(rows, cols)

    for row in range(rank):
        # Check if element at [row][row] is nonzero
        if matrix[row][row] != 0:
            # Make all elements below this element 0
            for col in range(row + 1, rows):
                factor = matrix[col][row] / matrix[row][row]
                for i in range(cols):
                    matrix[col][i] -= factor * matrix[row][i]
        else:
            # Find a row below with nonzero element at [row][row] and swap
            reduce = True
            for i in range(row + 1, rows):
                if matrix[i][row] != 0:
                    matrix[row], matrix[i] = matrix[i], matrix[row]
                    reduce = False
                    break
            # If no such row found, reduce rank
            if reduce:
                rank -= 1
                # Move the current row to the bottom
                matrix[row], matrix[-1] = matrix[-1], matrix[row]

    return rank

# Get user input for matrix
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))
matrix = []
print("Enter the matrix elements row by row:")
for i in range(rows):
    row = list(map(int, input().split()))
    matrix.append(row)

# Calculate and print the rank
rank = matrix_rank(matrix)
print("Rank of the matrix is:", rank)
```
