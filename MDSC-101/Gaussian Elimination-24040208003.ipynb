```python
def input_matrix_and_vector():
    n = int(input("Enter the number of variables (i.e., the order of the square matrix): "))

    # Input the coefficient matrix
    print("Enter the matrix (row by row, values space-separated):")
    A = []
    for i in range(n):
        row = list(map(float, input(f"Enter row {i+1}: ").split()))
        if len(row) != n:  # Check if it's a square matrix
            raise ValueError("Each row must contain exactly {} elements (square matrix).".format(n))
        A.append(row)

    # Input the right-hand side vector
    print("Enter the right-hand side vector (space-separated):")
    b = list(map(float, input().split()))
    if len(b) != n:
        raise ValueError("The right-hand side vector must contain exactly {} elements.".format(n))

    return A, b

def gaussian_elimination(A, b):
    n = len(b)

    # Augment the matrix A with vector b
    for i in range(n):
        A[i].append(b[i])

    # Forward Elimination
    for i in range(n):
        # Make the diagonal element 1 and scale the row accordingly
        pivot = A[i][i]
        if pivot == 0:
            raise ValueError("Matrix is singular or requires row exchanges.")
        for j in range(i, n+1):
            A[i][j] /= pivot

        # Eliminate the below rows
        for k in range(i+1, n):
            factor = A[k][i]
            for j in range(i, n+1):
                A[k][j] -= factor * A[i][j]

    # Back Substitution
    x = [0 for _ in range(n)]
    for i in range(n-1, -1, -1):
        x[i] = A[i][n]
        for j in range(i+1, n):
            x[i] -= A[i][j] * x[j]

    return x

# Main Program
try:
    A, b = input_matrix_and_vector()  # User inputs the matrix and vector
    solution = gaussian_elimination(A, b)

    # Print the solution in the form xi = value
    for i, val in enumerate(solution):
        print(f"x{i+1} = {val:.4f}")

except ValueError as e:
    print(e)
```
