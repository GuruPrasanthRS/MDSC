1.   Read two vectors of user defined dimension and perform their inner product.

```python
def dot_product(vector1, vector2):
    # Check if both vectors have the same length
    if len(vector1) != len(vector2):
        raise ValueError("Vectors must be the same length")

    # Initialize the dot product to 0
    dot_product_result = 0

    # Loop through both vectors and calculate the dot product
    for i in range(len(vector1)):
        dot_product_result += vector1[i] * vector2[i]

    return dot_product_result

# Collecting input from the user
vector1 = list(map(int, input("Enter the first vector: ").split()))
vector2 = list(map(int, input(f"Enter the second vector with length {len(vector1)}: ").split()))

# Calculating the dot product
try:
    result = dot_product(vector1, vector2)
    print(f"The dot product of {vector1} and {vector2} is {result}")
except ValueError as e:
    print(e)
```

2.   Read two matrices of user defined order. Check for their compatibility and perform matrix multiplication.



```python
def input_matrix(rows, cols):
    matrix = []
    print(f"Enter the elements of a {rows}x{cols} matrix row by row with spaces:")
    for i in range(rows):
        row = list(map(int, input().split()))
        if len(row) != cols:
            raise ValueError(f"Row {i+1} does not have {cols} elements")
        matrix.append(row)
    return matrix

def print_matrix(matrix):
    for row in matrix:
        print(" ".join(map(str, row)))

def matrix_multiply(A, B):
    # Get the dimensions of the matrices
    rows_A, cols_A = len(A), len(A[0])
    rows_B, cols_B = len(B), len(B[0])

    # Check for compatibility
    if cols_A != rows_B:
        raise ValueError("Number of columns of the first matrix must be equal to the number of rows of the second matrix")

    # Initialize the result matrix with zeros
    result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    # Perform the matrix multiplication
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]

    return result

# Get the order of the first matrix
rows_A = int(input("Enter the number of rows in the first matrix: "))
cols_A = int(input("Enter the number of columns in the first matrix: "))

# Get the first matrix from the user
matrix_A = input_matrix(rows_A, cols_A)

# Get the order of the second matrix
rows_B = int(input("Enter the number of rows in the second matrix: "))
cols_B = int(input("Enter the number of columns in the second matrix: "))

# Get the second matrix from the user
matrix_B = input_matrix(rows_B, cols_B)

try:
    # Perform matrix multiplication
    result_matrix = matrix_multiply(matrix_A, matrix_B)

    # Print the result
    print("The resulting matrix after multiplication is:")
    print_matrix(result_matrix)
except ValueError as e:
    print(e)
```
